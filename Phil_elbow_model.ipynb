```python
import os
import numpy as np
import cv2
from tqdm import tqdm
import tensorflow as tf
import matplotlib.pyplot as plt
from random import shuffle
import tflearn
from tflearn.layers.conv import conv_2d,max_pool_2d
from tflearn.layers.core import input_data,dropout,fully_connected
from tflearn.layers.estimator import regression

def predict():
    path=".//upload"

    convnet=input_data(shape=[None,50,50,1],name='input')
    model=tflearn.DNN(convnet,tensorboard_dir='log')
    model.load('my_model.tflearn')
    test_data=[]
    img_data=cv2.imread('''sys argument to be passed here''',cv2.IMREAD_GRAYSCALE);
    img_data=cv2.resize(img_data,(50,50));
    test_data.append([np.array(img_data),1])
    test_data.append([np.array(img_data),2])
    for num,data in enumerate(test_data):
        img_num=data[1]
        img_data=data[0]
        orig=img_data
        data=img_data.reshape(50,50,1)
        model_out=model.predict([data])[0]
        if np.argmax(model_out)==1:
            str_label='Fracture'
        else:
            str_label='Not Fracture'
        plt.imshow(orig,cmap='gray')
        plt.title(str_label)
        plt.gca().set_axis_off()
        plt.show()
        break;
predict()
```
